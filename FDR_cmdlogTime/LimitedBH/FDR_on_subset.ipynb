```python
import pandas as pd
import numpy as np
from statsmodels.stats import multitest
FN_GENES_OF_INTEREST = "genesOfInterest.txt"
FN_GENES_OF_INTEREST_REDUCED = "genesOfInterestReduced.txt"
DESEQ2_RESULTS = "rnd.deSeq2.noNA.tsv"
```

```python
with open (FN_GENES_OF_INTEREST, 'r') as f:
    genes_of_interest = [line.strip() for line in f][1:]
df = pd.read_csv(DESEQ2_RESULTS, sep="\t", index_col=0)
df['fdr'] = multitest.multipletests(df['pval'].to_numpy(), method='fdr_bh')[1]
df.to_csv('rnd.deSeq2_fdr.tsv', sep='\t')
goi_subset = [g for g in genes_of_interest if g in df.index]
df_goi = df.loc[goi_subset,:]
```

```python
#type (df_goi['pval'])
df_goi['fdr'] = multitest.multipletests(df_goi['pval'].to_numpy(), method='fdr_bh')[1]
df_goi.head()
df_goi.to_csv('rnd.deSeq2_fdr_goi.tsv', sep='\t')
df_goi
```

```python
with open (FN_GENES_OF_INTEREST_REDUCED, 'r') as f:
    genes_of_interest_reduced = [line.strip() for line in f][1:]
goir_subset = [g for g in genes_of_interest_reduced if g in df.index]
df_goir = df.loc[goir_subset,:]
```

```python
df_goir['fdr'] = multitest.multipletests(df_goir['pval'].to_numpy(), method='fdr_bh')[1]
df_goir.to_csv('rnd.deSeq2_fdr_goir.tsv', sep='\t')
df_goir
```
